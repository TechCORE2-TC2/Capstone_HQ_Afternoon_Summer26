# **TASK 2: ADD DATA TO SONGS WITH LOOPS & CONDITIONALS**

--

Two additional pieces of data need to be added to the songs:
- *Era*: A string label describing each song's era based on the year realeased. Options are "contemporary" (2015-present), "classic" (2000-2014), and "oldie" (< 2000)
- *Popularity*: A string label for each song's popularity based on streams. Options are "top hits" (> 1,000,000 ), "well known" (100,000 - 999,999), and "underground" (< 100,000)

````python
''' 2) FOR-LOOP AND CONDITIONAL STATEMENTS FOR ERA / POPULARITY '''

# FILL IN CODE BELOW
for song in song_list:
    # check year to add an era label
    year = song[4]
    streams = song[3]

    if year >= 2015:
        song.append("contemporary")
    elif year >= 2000:
        song.append("classic")
    else:
        song.append("underground")

    # check streams to add popularity label
    if streams >= 1000000:
        song.append("top hits")
    elif streams >= 100000:
        song.append("well known")
    else:
        song.append("underground")
    
    # check song to make sure it worked
    print(song)
````

# **TASK 3: DEFINE FUNCTION FOR CREATING SONG DATABASE**

--

The song_list now must be put into a *Pandas Dataframe* to be used on the website. This is a common Python data format with widespread applications. You don't need to know much about how it works now, your current skills can be used to make the dataframe.

### 3a)

Complete the function for creating the playlist database and the for-loop

````python
''' 3a) DEFINE FUNCTION FOR CREATING PANDAS PLAYLIST'''

# packages
import pandas as pd
import numpy as np

# FILL IN CODE BELOW
# we need to pass song_list as the argument for the function, but use a different variable name
def create_panda_playlist(songs):

    # Create a dataframe for song database
    playlist_df = pd.DataFrame(columns=[
        'title',
        'artists',
        'youtube_url',
        'streams',
        'date_released',
        'explicit',
        'era',
        'popularity'
    ])

    # loop through all songs in LoL and append to pandas dataframe
    for song in songs:
        playlist_df.loc[len(playlist_df)]= song
    
    # display playlist to ensure correctness
    display(playlist_df)
    return playlist_df
    
````

### 3b)

Initialize a new variable called *playlist* by calling the create_panda_playlist function
